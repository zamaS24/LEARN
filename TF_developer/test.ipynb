```python


embedding_dim = 16
vocab_size = 1000
max_length = 20

model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length)
])

model.summary()

print('input tensor shape : ', model.layers[0].input.shape) 

# > input tensor shape : (None, 20) 
# this means that we can input (batch_size, 20)
# since the sequence length is just 20 

print('ouptus tensor shape :', model.layers[-1].output.shape)
```

```python
import os

model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 300x300 with 3 bytes color
    # This is the first convolution

    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(300, 300, 3), name='conv2D_1'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # The second convolution
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # The third convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # The fourth convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # The fifth convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),

    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),

    # Only 1 output neuron. It will contain a value from 0-1 where 0 for 1 class ('horses') and 1 for the other ('humans')
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.summary()
```

```python
n = len(model.layers) - 1

new_model = tf.keras.Model(
    inputs  = model.layers[0].input, 
    outputs = model.layers[-1].output
)


# # new_model.summary()
# for layer in new_model.layers: 
#     print('layer type ', type(layer)) 

vocab_size = 1000
embedding_dim = 16

model2 = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=20),

    # I guess cuz it's fuly connected right ?

    # add an LSTM layer 
    tf.keras.layers.LSTM(units=32, return_sequences=False, return_state=False),

    tf.keras.layers.Dense(1, activation='sigmoid')
])


model2.compile(
    optimizer='adam',
    loss='binary_corssentropy',
    metrics=['accuracy']
)

# layers
layers = model2.layers
print('number of layers is : ', len(layers))

# lstm layer
lstm_layer = layers[1]
print('**** debug lstm layers ****')
print('lstm layer ',lstm_layer)
print('embedding layer output :', layers[0].output.shape)
print('lstmlayer input :', lstm_layer.input.shape)
print('lstm layer output :', lstm_layer.output.shape)
print()


# print the model summary 
model2.summary()

```
