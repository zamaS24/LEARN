```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# generating random data_points 
x_train = np.linspace(0,6)
y_train = 2 * x_train -2


plt.plot(x_train, y_train, 'o')
plt.show()


W = np.array([0, 0], dtype='float')
print(f'x_train shape : ', x_train.shape)
```

```python
# compute hw aka ^y 
def forward(X, W): 
    W = W.reshape(2,-1)
    X = np.stack((np.ones_like(X), X), axis=-1)

    return  np.dot(X, W)
  
# Mean Square Error Loss
def loss(Y, hw):
    l =  0.5 * ((Y - hw).mean())
    
    return l 
```

```python
W = W 
X = x_train 
Y = y_train

# training parameters
epochs = 100000
alpha = 0.005

# This is for debugiing purposes
use_DEBUG = False

losses= []
for epoch in range(epochs):

    hw = forward(X,W)
    hw = hw.reshape(-1)
    
    # for debugging purposes
    if (use_DEBUG): 
        print('---------- iter ',epoch, '-------------')
        print('shape of X is : ', X.shape)
        print('Shape of Y is : ', Y.shape)
        print('Shape of hw   : ', hw.shape)
        print('gradient 0 == ', gradient0)
        print('gradient 1 == ', gradient1)
    
    # calculate gradients
    gradient0 = (hw - Y).mean()     # dcost(w)/dW0
    gradient1 = (X * (hw - Y)).mean() # # dcost(w)/dW1

    # updating gradients
    W[0] = W[0] - alpha* gradient0
    W[1] = W[1] - alpha *gradient1

    if (epoch % 10 == 0): 
        print('epoch : ', epoch)
        current_loss = loss(Y, hw)
        print('loss : ', current_loss )
        losses.append(current_loss)
        print()
```
