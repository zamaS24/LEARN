```python

class Net(nn.Module): 
    
    # Override 
    def __init__(self): 
        super(Net, self).__init__()

        self.fc1 = nn.Linear(in_features=2, out_features=4)
        self.fc2 = nn.Linear(in_features=4, out_features=2)
        self.out = nn.Linear(in_features=2, out_features=1)

        # convolutions
        self.conv1 = nn.Conv2d(1,2,3)
        self.conv2 = nn.Conv2d(1,2,3)

        # max_pooling 
    


    # Override 
    def forward(self, x:torch.tensor) -> torch.tensor: 
        
        x = nn.ReLU(self.fc1(x))
        x = nn.ReLu(self.fc2(x))
        x = nn.ReLu(self.out(x))

        return x



model = Net() 

from os import path 
from os import getcwd 

# prepare the path 
dirname = getcwd()
filename = 'model.pth'
model_path  = path.join(dirname, filename)

torch.save(model, model_path)
    

# to train the model you need a dataset, and a data loader , and you can perform some transfomrations too 
# we should've been in february now, and started doing the shit of the two chapters man 
```
