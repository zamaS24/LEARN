```python
import torch
import torch.nn as nn 
import torchvision 
import torchvision.datasets as datasets

from torch.utils.data import DataLoader
from torchvision.transforms import transforms
import numpy as np 
```

```python
# create and prepare the datasets 
root ='./data'

# 1. Download the dataset 
# I already downloaded the dataset so no need to see 
train_dataset = datasets.CIFAR10(root=root, train=True, download=True)
val_dataset = datasets.CIFAR10(root=root, train=False, download=True)


# 2. Load the dataset 

# first define the transfomrations 
mean = np.array([0.5, 0.5, 0.5])
std = np.array([0.25, 0.25, 0.25])

transformations = transforms.Compose([
    transforms.Resize((224,224)), 
    transforms.ToTensor(), 
    transforms.Normalize(mean, std)
])

# use ImageLoader 
train_dataset = datasets.ImageFolder(root='./data', transform=transformations) # I think it should be /data/train
val_dataset = datasets.ImageFolder(root='./data', transform=transformations) # I think it should be /data/val



# 3. Create data loader to iterate and batches 
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(val_dataset, batch_size=42, shuffle=True)




print('len of train_loader ', len(train_loader))
print('len of valid loader', len(valid_loader))
```
