> :warning: Be careful wth memory location if you have b = a.numpy() and then you change a, you will also see that b has changes, since they point to the same memory location

```python
import numpy as np 

a = np.ones(5)
print(a)

b = torch.from_numpy(a)
print(b)

a += 1 

print('a after modification ', a)
print('b after modification ', b)
# here also the b will be changes lol 


if (torch.cuda.is_available()): 
    print('available')
    device = torch.device('cuda')
    x = torch.ones(5, device = device) # this will make it on the gpu 
    y = torch.ones(5)
    y = y.to(device) 
    z = x + y # this might be faster since it's on the gpu 

    z.numpy() # this will return an error since numpy can handle only CPU tensors
    # you cannot convert a gpu tensor to cpu 

    z = z.to('cpu') # this will return it to cpu    
else: 
    print('not availabel')



```
