```python
import pandas as pd
pd.set_option("display.max_rows", 5)
reviews = pd.read_csv("datasets/winemag-data-130k-v2.csv", index_col=0)
# rename columns
reviews.rename(columns={'points':'score'})
```

### Combining `concat()` `join()` `merge()`

```python

canadian_youtube = pd.read_csv('datasets/youtube_trending/CAvideos.csv' )
british_youtube = pd.read_csv('datasets/youtube_trending/GBvideos.csv' )

pd.concat([british_youtube, canadian_youtube])
```

lsuffix and rsuffix are important so we can see these attributes in the tabels if they provide from uk or canada

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])


left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
