9. Create a scatter plot to visualize the relationship between the total bill and the tip amount. Customize the plot with labels and title.

```python
# sns.scatterplot(x='sepal_length', y='sepal_width', data=data,hue="species")

sns.scatterplot(x='total_bill', y='tip', data=df)
plt.xlabel('total bill')
plt.ylabel('tip mount') 
plt.title('Scatter:  total bill vs tip ')
plt.show()
```

 10. Generate a box plot to compare the distribution of tips for different party sizes. Customize the appearance of the plot.

```python
# sns.boxplot(x='species', y='sepal_width', data=data, palette="Set2")
sns.boxplot(x='size', y='tip', data=df, palette="Set2")
plt.title('Box Plot: tip Amount by party size')
plt.show()
```

- Subplot 1: Scatter Plot  
    Generate a scatter plot to visualize the relationship between 'total_bill' and 'tip'. Customize the appearance of the scatter plot.

    Create a point plot to show the average total bill for each day of the week. Customize the colors and labels.

- Subplot 4: Violin Plot for Tip Distribution  
    Create a violin plot to visualize the distribution of tips for each day. Customize the appearance of the violin plot.   
Use the subplot2grid function  

```python
plt.figure(figsize=(20,10), dpi=100)

# Scatter plot
def graph1(): 
    sns.scatterplot(x='total_bill', y='tip', data=df)
    pass

# point plot
def graph2(): 
    sns.pointplot(x='day', y='total_bill', data=df, palette='Set2')
    pass

# box plot for launch and dinner
def graph3():
    sns.boxplot(x='time', y='total_bill', data=df, palette='Set3')
    pass

# viloin plot for tip distribution 
def graph4(): 
    sns.violinplot(x='day', y='tip', data=df, palette='Set1')
    pass


axes1 = plt.subplot2grid((2, 2), (0, 0))
graph1()

axes2 = plt.subplot2grid((2, 2), (0, 1))
graph2()

axes3 = plt.subplot2grid((2, 2), (1, 0))
graph3()

axes4 = plt.subplot2grid((2, 2), (1, 1))
graph4()

```
