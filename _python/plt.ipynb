```python
import matplotlib.pyplot as plt 
import numpy as np 
```

# 1. Pyplot

```python
# some dict fonts
font1 = {
    'family':'serif',
    'color':'blue',
    'size':20
}

font2 = {
    'family':'serif',
    'color':'darkred',
    'size':15
}

xpoints = np.array([-1,0, 6, 7])
ypoints = np.array([2,1, 2, 3])

plt.plot(xpoints, ypoints)



plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.show()
```
